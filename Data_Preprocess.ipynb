```python
#!/usr/bin/env python3
# -*- coding:utf8 -*-
# @TIME  :2018/9/17 9:02
# @Author:dazhan
# @File  :copyfiles2dir.py


#这一步就是复制出来，防止污染源数据

import os
import shutil

source_path = os.path.abspath(r'/home/amy/Desktop/data/train/2/')
target_path = os.path.abspath(r'/home/amy/Desktop/data/train/nevus/')

if not os.path.exists(target_path):
    os.makedirs(target_path)

if os.path.exists(source_path):
    for root, dirs, files in os.walk(source_path):
        for file in files:
            src_file = os.path.join(root, file)
            shutil.copy(src_file, target_path)
            print(src_file)

print('copy files finished!')
```

```python
#把bmp格式的文件转成jpg

import os
from PIL import Image

# bmp to jpg
def bmpToJpg(file_path):
    count = 0
    for fileName in os.listdir(file_path):
        # print(fileName)
        newFileName = str(count)+".jpg"
        print(newFileName)
        im = Image.open(file_path+"/"+fileName)
        im.save(file_path+"/"+newFileName)
        count += 1


# Delete the original bitmap
def deleteImages(file_path, imageFormat):
    command = "del "+file_path+"/*."+imageFormat
    os.system(command)


def main():
    file_path = "/home/amy/Desktop/data/train/nevus/"
    bmpToJpg(file_path)
    deleteImages(file_path, "bmp")
```

```python
main()
```

```python
#将数据分为训练集和测试集

import os, random, shutil
def move(fileDir, tarDir, rate):
    pathDir = os.listdir(fileDir)
    filenumber = len(pathDir)
    picknumber = int(filenumber*rate)
    sample = random.sample(pathDir, picknumber)
    for name in sample:
        shutil.move(os.path.join(fileDir,name), os.path.join(tarDir, name))
    return

def split(typeadd, rate):
    for root, dirs, files in os.walk(os.path.abspath(r'/home/amy/Desktop/data/train/')):
        for d in dirs:
            new = os.path.join(typeadd, d)
            if not os.path.exists(new):
                os.makedirs(new)
            move(os.path.join(root, d), new, rate)


split(os.path.abspath(r'/home/amy/Desktop/data/valid/'),0.2)
print('=======valid finish=======')
split(os.path.abspath(r'/home/amy/Desktop/data/test/'),0.15)
print('========test finish=======')
```
